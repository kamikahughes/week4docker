```python
#Start training/modeling
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
```

```python
y = df['Class']
X = df[['Aspect_Ratio', 'Uniformity', 'Elongation','Solidity']]
```

```python
X_train,X_test, y_train, y_test= train_test_split(X, y, test_size=0.25,random_state=3)
```

```python
lgreg= LogisticRegression(max_iter=200) 
lgreg.fit(X_train, y_train.ravel()) # ravel() is required here, system will give error and suggest using ravel()
Threshold=0.5
y_pred=lgreg.predict(X_test)
```

```python
LGRG= pd.DataFrame({"y_test": y_test.ravel(), "Y-pred": y_pred.ravel()})
LGRG.head(10)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
lreg.fit(X_train, y_train)
y_pred= lreg.predict(X_test)
```

```python
result = pd.DataFrame({"y_test":y_test.ravel(),"y_pred":y_pred.ravel()})
result.head(10)
```

```python
mse = mean_squared_error(y_test,y_pred)
print("MSE is ",mse)
```

```python
print(lreg.coef_, lreg.intercept_)
```

```python
from sklearn.tree import DecisionTreeClassifier
#Criterion parameter can be entropy, gini, logloss, with default = gini.
dtree = DecisionTreeClassifier(criterion='gini',random_state=1) 
dtree.fit(X_train,y_train)
y_pred = dtree.predict(X_test)
result= pd.DataFrame({"y_test": y_test.ravel(), "Y-pred": y_pred.ravel()})
result.head(10)
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```
